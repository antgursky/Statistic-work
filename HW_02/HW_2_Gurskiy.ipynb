# Ответы на вопросы

---

### 1. Что такое доверительный интервал?  
Доверительный интервал — это диапазон значений, который с заданной вероятностью (уровнем доверия, например, $95\%$) содержит истинный параметр генеральной совокупности. Формально: если построить множество интервалов по разным выборкам, то доля интервалов, содержащих истинный параметр, стремится к уровню доверия.

---

### 2. Чем асимптотический доверительный интервал отличается от неасимптотического?  
- **Асимптотический**:  
  Корректность гарантируется только при $n \to \infty$ (основан на ЦПТ). Пример:  
  $$\hat{\theta} \pm z_{\alpha/2} \cdot \text{SE}.$$  
- **Неасимптотический**:  
  Точен для конечных $n$, но требует строгих условий (например, нормальности данных). Пример: $t$-интервал для малых выборок.

---

### 3. Что больше: квантиль стандартного нормального распределения уровня $0.05$ или уровня $0.1$?  
- Левосторонние квантили:  
  $z_{0.05} \approx -1.645$, $z_{0.1} \approx -1.282$ ⇒ $z_{0.05} < z_{0.1}$.  
- Правосторонние квантили:  
  $z_{0.95} = 1.645 > z_{0.9} = 1.282$.  
**Ответ**: Зависит от контекста. Если речь о нижних хвостах, квантиль $0.05$ меньше.

---

### 4. Распределение $Z = \sqrt{2}X_1 / \sqrt{X_2^2 + X_3^2}$  
- $X_2^2 + X_3^2 \sim \chi^2_2$ ⇒ $\sqrt{X_2^2 + X_3^2} \sim \chi_2$ (хи-распределение).  
- Числитель: $\sqrt{2}X_1 \sim \mathcal{N}(0, 2)$.  
- Перепишем:  
  $$  
  Z = \frac{\sqrt{2}X_1}{\sqrt{X_2^2 + X_3^2}} = \frac{X_1}{\sqrt{\frac{X_2^2 + X_3^2}{2}}} \sim t_2,  
  $$  
  где $t_2$ — распределение Стьюдента с $2$ степенями свободы.

---

### 5. Плюсы, недостатки и популярность бутстрепа  
**Плюсы**:  
- Не требует предположений о распределении данных.  
- Применим для любых статистик:  
  $$\hat{\theta}_{\text{boot}} = \frac{1}{B} \sum_{i=1}^B \hat{\theta}_i^*.$$  
- Универсален для малых выборок.  

**Недостатки**:  
- Вычислительная сложность: $O(B \cdot n)$, где $B$ — число бутстреп-повторов.  
- Смещение для некоторых оценок (например, экстремальных порядковых статистик).  
- Не работает при зависимых данных (требует модификаций).  

**Популярность**: Обходит сложности аналитического вывода, особенно для нестандартных задач.

# Ответ на упражнение 2

---

**Условие задачи**  
Пусть $  X  $ — случайная величина с $ \mathbb{E}X = a $ и $ \text{Var}(X) = \sigma^2 $. С помощью неравенства Чебышёва оцените вероятности:  
1. $  P(X \in [a - 2\sigma, a + 2\sigma])  $.  
2. $  P(X \in [a - 3\sigma, a + 3\sigma])  $.  
Сравните с нормальным распределением $ \mathcal{N}(0,1) $ и сделайте вывод.

---

**Решение**  
1. **Неравенство Чебышёва**:  
   Для любого $  k > 0  $:  
   $$
   P(|X - a| \geq k\sigma) \leq \frac{1}{k^2} \implies P(|X - a| \leq k\sigma) \geq 1 - \frac{1}{k^2}.
   $$  

2. **Оценка для $  k = 2  $**:
   $$
   P(|X - a| \leq 2\sigma) \geq 1 - \frac{1}{2^2} = \frac{3}{4} = 0.75.
   $$  

3. **Оценка для $  k = 3  $**:
   $$
   P(|X - a| \leq 3\sigma) \geq 1 - \frac{1}{3^2} = \frac{8}{9} \approx 0.8889.
   $$  

4. **Сравнение с $ \mathcal{N}(0,1) $**:
   - Для нормального распределения:  
     - $  P(|Z| \leq 2) \approx 0.9545  $,  
     - $  P(|Z| \leq 3) \approx 0.9973  $.  
   - Оценки Чебышёва ($  0.75  $ и $  0.8889  $) значительно ниже реальных значений для нормального распределения.

---

**Вывод**  
Неравенство Чебышёва дает универсальные, но консервативные оценки вероятности. Для нормального распределения фактические вероятности намного выше, чем оценки Чебышёва. Это показывает, что:  
- Неравенство Чебышёва применимо к любым распределениям с конечной дисперсией,  
- Для «хороших» распределений (например, нормального) реальные вероятности отклонений от среднего существенно лучше оценок Чебышёва.

# Ответ на упражнение 3

---

### **Часть 1: Точный доверительный интервал через максимум выборки**

**Используемая статистика**: $ \hat{\theta}_1 = x_{(n)} = \max\{x_1, \ldots, x_n\} $.  

1. **Распределение $ x_{(n)} $**  
   Для $ X \sim U[0, \theta] $ функция распределения максимума:  
   $$
   F_{\hat{\theta}_1}(u) = P(x_{(n)} \leq u) = \left(\frac{u}{\theta}\right)^n, \quad u \in [0, \theta].
  $$

2. **Построение интервала**  
   Для уровня доверия $ 1 - \alpha $ найдем квантили $ c_1, c_2 $:  
   $$
   P(c_1 \leq \hat{\theta}_1 \leq c_2) = 1 - \alpha.
   $$
   Решаем уравнения:  
   $$
   \left(\frac{c_1}{\theta}\right)^n = \frac{\alpha}{2}, \quad \left(\frac{c_2}{\theta}\right)^n = 1 - \frac{\alpha}{2}.
   $$
   Отсюда:  
   $$
   c_1 = \theta \left(\frac{\alpha}{2}\right)^{1/n}, \quad c_2 = \theta \left(1 - \frac{\alpha}{2}\right)^{1/n}.
  $$
   Перепишем интервал для $ \theta $ через $ \hat{\theta}_1 $:  
   $$
   \theta \geq \frac{\hat{\theta}_1}{\left(1 - \frac{\alpha}{2}\right)^{1/n}}, \quad \theta \leq \frac{\hat{\theta}_1}{\left(\frac{\alpha}{2}\right)^{1/n}}.
  $$
   **Итоговый интервал**:  
   $$
   \left[ \frac{\hat{\theta}_1}{\left(1 - \frac{\alpha}{2}\right)^{1/n}}, \ \frac{\hat{\theta}_1}{\left(\frac{\alpha}{2}\right)^{1/n}} \right].
  $$

---

### **Часть 2: Асимптотический доверительный интервал через удвоенное среднее**

**Используемая статистика**: $ \hat{\theta}_2 = 2\bar{x} $.  

1. **Свойства $ \hat{\theta}_2 $**  
   - Матожидание: $ \mathbb{E}\hat{\theta}_2 = \theta $.  
   - Дисперсия: $ \text{Var}(\hat{\theta}_2) = \frac{\theta^2}{3n} $.  

2. **Асимптотическая нормальность**  
   По ЦПТ:  
   $$
   \sqrt{n} \left(\hat{\theta}_2 - \theta\right) \xrightarrow{d} \mathcal{N}\left(0, \frac{\theta^2}{3}\right).
  $$
   Стандартная ошибка (SE):  
   $$
   \text{SE} = \frac{\theta}{\sqrt{3n}}.
  $$

3. **Построение интервала**  
   Заменяем $ \theta $ на оценку $ \hat{\theta}_2 $ в SE:  
   $$
   \text{SE} \approx \frac{\hat{\theta}_2}{\sqrt{3n}}.
   $$
   **Асимптотический интервал**:  
   $$
   \hat{\theta}_2 \pm z_{\alpha/2} \cdot \frac{\hat{\theta}_2}{\sqrt{3n}},
   $$
   где $ z_{\alpha/2} $ — квантиль $ \mathcal{N}(0, 1) $.  

---

### **Сравнение**  
- **Точный интервал**: Использует распределение максимума, корректен для любых $ n $.  
- **Асимптотический интервал**: Точен только при $ n \to \infty $, требует замены $ \theta $ на оценку в SE.  

**Вывод**: Для малых выборок предпочтителен точный интервал через $ \hat{\theta}_1 $, для больших — асимптотический через $ \hat{\theta}_2 $.

# Ответ на упражнение 4
### **1. Дисперсия исходного распределения**  
Для $X \sim U[0, \theta]$:  
$$
\text{Var}(X) = \frac{\theta^2}{12}.
$$

---

### **2. Дисперсия новой выборки**  
Новое распределение: $Y \sim U[0, 2\bar{x}]$, где $\bar{x} = \frac{1}{n}\sum_{i=1}^n x_i$.  
Дисперсия $Y$:  
$$
\text{Var}(Y) = \frac{(2\bar{x})^2}{12} = \frac{\bar{x}^2}{3}.
$$

---

### **3. Сравнение дисперсий**  
- **Ожидаемая дисперсия новой выборки**:  
  Так как $\mathbb{E}[\bar{x}] = \frac{\theta}{2}$ (среднее равномерного распределения на $[0, \theta]$),  
  $$
  \mathbb{E}[\text{Var}(Y)] = \frac{\left(\mathbb{E}[\bar{x}]\right)^2}{3} = \frac{\left(\frac{\theta}{2}\right)^2}{3} = \frac{\theta^2}{12} = \text{Var}(X).
  $$  

- **Фактическая дисперсия**:  
  Зависит от конкретной реализации $\bar{x}$. Например, если $\bar{x} \neq \frac{\theta}{2}$, то:  
  $$
  \text{Var}(Y) = \frac{\bar{x}^2}{3} \neq \frac{\theta^2}{12}.
  $$  

---

### **4. Вывод**  
- **В среднем** (по всем возможным выборкам) дисперсия новой выборки совпадает с исходной.  
- **На практике** из-за случайности $\bar{x}$ дисперсия новой выборки может отличаться от исходной.  
### **Итог**:
Параметрический бутстреп с оценкой
$2\bar{x}$ сохраняет ожидаемую дисперсию, но в конкретных реализациях возможны отклонения. Это подчеркивает важность учета стохастической природы оценок в бутстрепе.

# Ответ на упражнение 5

---

### **Шаги выполнения эксперимента**

1. **Генерация параметров**:  
   - $\theta \sim U[10, 20]$, $\sigma \sim U[1, 2]$.  





2. **Генерация выборки**:  
   - $X \sim \mathcal{N}(\theta, \sigma^2)$, $n = 50$:  


````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t

# Время исполнения кода займет ~8 минут

# Настройка стиля графиков
sns.set(style="whitegrid")
plt.rcParams["font.family"] = "DejaVu Sans"

# 1. Параметры эксперимента
np.random.seed(42)
n_experiments = 10000
n_sample = 50
alpha = 0.05
bootstrap_samples = 1000

# 2. Функции для построения интервалов
def theoretical_ci(sample, alpha):
    n = len(sample)
    mean = np.mean(sample)
    std = np.std(sample, ddof=1)
    t_crit = t.ppf(1 - alpha/2, df=n-1)
    margin = t_crit * std / np.sqrt(n)
    return (mean - margin, mean + margin)

def parametric_bootstrap(sample, alpha, n_bootstrap):
    means = []
    n = len(sample)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    for _ in range(n_bootstrap):
        bootstrap_sample = np.random.normal(sample_mean, sample_std, n)
        means.append(np.mean(bootstrap_sample))
    return np.quantile(means, [alpha/2, 1 - alpha/2])

def nonparametric_bootstrap(sample, alpha, n_bootstrap):
    means = []
    n = len(sample)
    for _ in range(n_bootstrap):
        bootstrap_sample = np.random.choice(sample, size=n, replace=True)
        means.append(np.mean(bootstrap_sample))
    return np.quantile(means, [alpha/2, 1 - alpha/2])

# 3. Проведение эксперимента
results = {
    "theoretical": {"coverage": 0, "lengths": []},
    "parametric": {"coverage": 0, "lengths": []},
    "nonparametric": {"coverage": 0, "lengths": []}
}

for _ in range(n_experiments):
    # Генерация параметров
    true_theta = np.random.uniform(10, 20)
    true_sigma = np.random.uniform(1, 2)

    # Генерация выборки
    sample = np.random.normal(true_theta, true_sigma, n_sample)

    # Построение интервалов
    # Теоретический
    ci_theory = theoretical_ci(sample, alpha)
    results["theoretical"]["coverage"] += (ci_theory[0] <= true_theta <= ci_theory[1])
    results["theoretical"]["lengths"].append(ci_theory[1] - ci_theory[0])

    # Параметрический бутстреп
    ci_param = parametric_bootstrap(sample, alpha, bootstrap_samples)
    results["parametric"]["coverage"] += (ci_param[0] <= true_theta <= ci_param[1])
    results["parametric"]["lengths"].append(ci_param[1] - ci_param[0])

    # Непараметрический бутстреп
    ci_nonparam = nonparametric_bootstrap(sample, alpha, bootstrap_samples)
    results["nonparametric"]["coverage"] += (ci_nonparam[0] <= true_theta <= ci_nonparam[1])
    results["nonparametric"]["lengths"].append(ci_nonparam[1] - ci_nonparam[0])

# 4. Анализ результатов
for method in results:
    results[method]["coverage"] /= n_experiments

# 5. Визуализация
# Гистограммы длин интервалов
plt.figure(figsize=(12, 6))
plt.hist(results["theoretical"]["lengths"], bins=50, alpha=0.5, label="Теоретический")
plt.hist(results["parametric"]["lengths"], bins=50, alpha=0.5, label="Параметрический")
plt.hist(results["nonparametric"]["lengths"], bins=50, alpha=0.5, label="Непараметрический")
plt.xlabel("Длина интервала")
plt.ylabel("Частота")
plt.title("Распределение длин доверительных интервалов")
plt.legend()
plt.show()

# Boxplot длин интервалов
plt.figure(figsize=(10, 6))
data = [
    results["theoretical"]["lengths"],
    results["parametric"]["lengths"],
    results["nonparametric"]["lengths"]
]
plt.boxplot(data, labels=["Теоретический", "Параметрический", "Непараметрический"])
plt.ylabel("Длина интервала")
plt.title("Сравнение длин доверительных интервалов")
plt.show()

# График покрытия
plt.figure(figsize=(10, 6))
methods = ["theoretical", "parametric", "nonparametric"]
coverages = [results[method]["coverage"] for method in methods]
plt.bar(methods, coverages, color=["blue", "green", "red"])
plt.axhline(0.95, color="black", linestyle="--", label="Ожидаемый уровень (95%)")
plt.ylim(0.9, 0.96)
plt.ylabel("Доля покрытия θ")
plt.title("Частота попадания θ в доверительные интервалы")
plt.legend()
plt.show()

# Вывод результатов
print("Результаты покрытия:")
for method in results:
    print(f"{method.capitalize()}: {results[method]['coverage']:.4f}")

print("\nСредние длины интервалов:")
for method in results:
    print(f"{method.capitalize()}: {np.mean(results[method]['lengths']):.4f}")
````

Формула:  
$$
\bar{x} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}},

где:
- $\bar{x}$ — выборочное среднее,
- $s$ — выборочное стандартное отклонение,
- $t_{\alpha/2, n-1}$ — квантиль t-распределения.

**Код**:  
```python
ci_theory = (mean - margin, mean + margin)
```


- **Дисперсия бутстреп-оценок**:  
  $$\text{Var}_{\text{параметрический}} = \frac{s^2}{n}, \quad \text{Var}_{\text{непараметрический}} = \frac{\hat{\sigma}^2_{\text{boot}}}{n}$$
