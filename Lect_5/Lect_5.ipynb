
views - сумма просмотров для одного пользователя

```python
views = np.int32(np.exp(np.random.normal(1, 1.5, size = 100000)))+1
                 
```

```python
_ = plt.hist(sorted(views)[:95000])

```

```python
# Бета распределение
alpha = 1000 #При большей выбореке
beta = 1000
ctr = np.random.beta(alpha,beta, size = 100000)
_= plt.hist(ctr)
```

```python
ctr
```

```python
click = np.random.binomial(views, ctr)
```

```python
click = plt.hist(sorted(click)[:95000])
```

```python
Проверяем наше респределение 
```

```python
def binomial1005(k):
    n = 10
    p = 1/2
    return binom(n,k)*(p**k)*(1-p)**(n-k)
```

```python
sum(binomial1005(i) for i in (0,1,2,8,9,10))
```
